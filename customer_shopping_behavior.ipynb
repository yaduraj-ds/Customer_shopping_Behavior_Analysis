```python
# Imputing missing values in Review Rating column with the median rating of the product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns.tolist()
```

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus

# --- CONFIGURATION ---
server = "localhost"
database = "customer_behavior"
driver = quote_plus("ODBC Driver 17 for SQL Server")

# Create connection string using Windows Authentication (Trusted Connection)
connection_string = f"mssql+pyodbc://@{server}/{database}?driver={driver}&trusted_connection=yes"

engine = create_engine(connection_string)

# --- EXECUTION ---
try:
    with engine.connect() as connection:
        print(f"âœ… Successfully connected to '{database}' on '{server}'!")

        # 1. Write DataFrame to SQL Server
        print("Writing data to SQL Server...")
        df.to_sql("customer", engine, if_exists="replace", index=False)
        print("Write complete.")

        # 2. Read back sample data
        print("Reading back top 5 rows:")
        result_df = pd.read_sql("SELECT TOP 5 * FROM customer;", engine)
        print(result_df)

except Exception as e:
    print("Connection failed:")
    print(e)
```
